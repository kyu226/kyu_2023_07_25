```python
import matplotlib.pyplot as plt
import yfinance as yf
data = yf.download('2303.TW',start='2022-01-01',end='2022-12-31').round(3)
tw2303 = data[['Adj Close']]
tw2303.columns=['']

# 
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.figure(figsize=(12, 6))  # 
plt.plot(tw2303.index, tw2303[''], label='', color='b')  # 
plt.title(' (2022)', fontsize=16)  # 
plt.xlabel('', fontsize=14)  # X
plt.ylabel('', fontsize=14)  # Y
plt.legend(loc='upper left')  # 
plt.grid(True)  # 

# 
plt.show()
```

```python
import pandas as pd
frame=pd.concat([tw2330,tw2303,tw2454,tw2317],
          axis=1).round(3)
frame

```

```python
frame1 = frame.pct_change()
frame2=frame1.fillna(0)
frame3=frame2.rename_axis('')
frame3
```

```python
#
frame2.corr()
```

```python
def fun1(s):
    max =s.max()*100
    min= s.min()*100
    return pd.Series([f'{max:.2f}%',f'{min:.2f}%'],index=['',''])

frame2.apply(fun1)
```

```python
from matplotlib.font_manager import FontProperties
import yfinance as yf
import matplotlib.pyplot as plt

# 
font = FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc', size=8)

# 
data = yf.download('2303.TW', start='2022-01-01', end='2022-12-31').round(3)
tw2303 = data[['Adj Close']]
tw2303.columns = ['']

# 
figure=plt.figure(figsize=(12, 6))  # 
plt.plot(tw2303.index, tw2303[''], label='', color='b')  # 
plt.title(' (2022)', fontsize=16,font= FontProperties)  # 
plt.xlabel('', fontsize=14)  # X
plt.ylabel('', fontsize=14)  # Y

# X
plt.gca().xaxis.set_ticklabels(tw2303.index.strftime('%Y-%m-%d'), fontproperties=font)

plt.legend(loc='upper left')  # 
plt.grid(True)  # 

# 
plt.show()
```
