```python
import matplotlib.pyplot as plt
import yfinance as yf
data = yf.download('2303.TW',start='2022-01-01',end='2022-12-31').round(3)
tw2303 = data[['Adj Close']]
tw2303.columns=['聯電']

# 设置中文字体为微软雅黑
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.figure(figsize=(12, 6))  # 设置图表大小
plt.plot(tw2303.index, tw2303['聯電'], label='聯電股價', color='b')  # 绘制折线图
plt.title('聯電股價走势 (2022年)', fontsize=16)  # 设置标题
plt.xlabel('日期', fontsize=14)  # 设置X轴标签
plt.ylabel('股價', fontsize=14)  # 设置Y轴标签
plt.legend(loc='upper left')  # 显示图例
plt.grid(True)  # 显示网格线

# 显示图表
plt.show()
```

```python
import pandas as pd
frame=pd.concat([tw2330,tw2303,tw2454,tw2317],
          axis=1).round(3)
frame

```

```python
frame1 = frame.pct_change()
frame2=frame1.fillna(0)
frame3=frame2.rename_axis('日期')
frame3
```

```python
#相關係數
frame2.corr()
```

```python
def fun1(s):
    max =s.max()*100
    min= s.min()*100
    return pd.Series([f'{max:.2f}%',f'{min:.2f}%'],index=['最高','最低'])

frame2.apply(fun1)
```

```python
from matplotlib.font_manager import FontProperties
import yfinance as yf
import matplotlib.pyplot as plt

# 设置中文字体
font = FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc', size=8)

# 下载股票数据
data = yf.download('2303.TW', start='2022-01-01', end='2022-12-31').round(3)
tw2303 = data[['Adj Close']]
tw2303.columns = ['聯電']

# 创建图表
figure=plt.figure(figsize=(12, 6))  # 设置图表大小
plt.plot(tw2303.index, tw2303['聯電'], label='聯電股價', color='b')  # 绘制折线图
plt.title('聯電股價走势 (2022年)', fontsize=16,font= FontProperties)  # 设置标题
plt.xlabel('日期', fontsize=14)  # 设置X轴标签
plt.ylabel('股價', fontsize=14)  # 设置Y轴标签

# 设置X轴刻度的字体属性
plt.gca().xaxis.set_ticklabels(tw2303.index.strftime('%Y-%m-%d'), fontproperties=font)

plt.legend(loc='upper left')  # 显示图例
plt.grid(True)  # 显示网格线

# 显示图表
plt.show()
```
