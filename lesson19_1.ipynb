```python
import pandas as pd 
import numpy as np

s1 = pd.Series([7.3,-2.5,3.4,1.5],index=['a','c','d','e'])
s2 = pd.Series([-2.1,3.6,-1.5,4,3.1],index=['a','c','e','f','g'])
display(s1)
display(s2)
#使用數學運算子的運算
display(s1+s2)
#使用實體方法的運算
s1.add(s2,fill_value=0)
```

```python
#sort_index()
s1 = pd.Series(range(4),index=['d','a','b','c'])
display(s1)
display(s1.sort_index())
```

```python
frame = pd.DataFrame(np.arange(8).reshape((2,4)),
                     index=['three','one'],
                     columns= ['d','a','b','c']
                     )
frame
```

```python
display(frame)
display(frame.sort_index())
display(frame.sort_index(axis='columns'))
```

```python
s3 = pd.Series([4,7,-3,2])
display(s3)
display(s3.sort_values())#由小到大
display(s3.sort_values(ascending=False)) #由大到小
```

```python
frame = pd.DataFrame({'b':[4,7,-3,2],
              'a':[0,1,0,1]
              })
display(frame)
display(frame.sort_values(by='b'))
display(frame.sort_values(by=['a','b']))
```

```python
#rank
obj = pd.Series([7,-5,7,4,2,0,4])
display(obj)
display(obj.rank(method='average'))
display(obj.rank(method='first'))
display(obj.rank(method='min'))
```

```python
df = pd.read_csv("學生成績.csv")
df.index=df['學號']
df1=df.drop(columns='學號')
df1['排名'] = df1['總分'].rank(method='min',ascending=False)
df1.sort_values(by='總分',ascending=False)
```
