```python
import pandas as pd

df = pd.read_csv("學生成績.csv")
df.index=df['學號']
df1 = df.drop(columns="學號")
df1
```

```python
df1['排名'] = df1['平均'].rank(method='min',ascending=False)
df2 = df1.sort_values(by='平均',ascending=False)
```

```python
#抓前10名
display(df2.iloc[0:10])
df2.query('排名<=10')
```

```python
df2.to_excel('學生成績單.xlsx',sheet_name="三年五班成績單")
```
