```python
def fun1(item):
    if item >= 80:
        return '優'
    elif item >= 70:
        return '甲'
    elif item >= 60:
        return '乙'
    else:
        return '丙'
    
students_frame['國文'].map(fun1)
```

```python
students_frame = frame[['國文','英文','數學','地理','歷史']]

def fun1(item):
    if item >= 80:
        return '優'
    elif item >= 70:
        return '甲'
    elif item >= 60:
        return '乙'
    else:
        return '丙'

students_frame.map(fun1)
```

```python
students_frame
```

```python
def fun2(s):
    return s.min()

students_frame.apply(fun2,axis=1)
```

```python
def fun3(s):
    return s[s<60].count()

students_frame.apply(fun3,axis=1)
```

```python
students_frame['不及格科目'] = students_frame.apply(fun3,axis=1)
students_frame
```

```python
def fun5(s):
    return pd.Series([s.max(),s.min(),s.median(),s.mean(),s.std()],
                     index=['最高分','最低分','中間值','平均','標準差']).round(2)

students_frame.apply(fun5)
```
